```python
#Having Fun with the Wikipedia library in Python. In this repo I show how you can use the wikipedia library to first read wikipeadia articles
#in a python IDE and then create word clouds from the output.
```

```python
#Import Libraries, make sure to install the wikipedia library if you haven't already. 
#Jupyter notebooks doesn't seem to come with these preinstalled.
#"pip/conda install wikipedia" from the CLI
#"pip/conda install wordcloud" from the CLI 
import pandas as pd 
import os
import numpy as np 
import wikipedia
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import io
#os.chdir('C:\\Users\\mike\\documents\\Python_Visualisations')
#os.getcwd()
```

```python
#The library is incredibly easy to use. Simply call the wikipedia library and function, pass a parameter which would be the keyword/title of the page you're looking for and you get an immediate result. 
#you get a summary of the wikipeadia page in the form of a string   
print(type(wikipedia.WikipediaPage(title = 'Chess Boxing').summary))
data_science= wikipedia.WikipediaPage(title = 'Chess Boxing').summary
data_science = data_science.replace('\n','').replace("\'","")
print(type(data_science))
print(data_science)
```

```python
#Now I convert the string I generated from wikipedia and convert it to an object 
df=pd.Series(data_science)
#then I pass the object into a dataframe at position [0:0]
df1=pd.DataFrame(df)
#then I rename the column Wiki_Text
df1.columns=['Wiki_Text']
df1.head()
```

```python
# Now we start creating a wordcloud 
text = df1.Wiki_Text[0]

# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=100, background_color="black").generate(text)
plt.figure()
#You've probably noticed the argument interpolation="bilinear" in the plt.imshow(). This is to make the displayed image appear more smoothly.
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
